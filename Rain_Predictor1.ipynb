```python
forecast_err=[]
for i,j in zip(df['RainTomorrow'][0:-2],df['RainToday'][1:-1]):
    forecast_err.append(i-j)
cc = pd.DataFrame(forecast_err,columns=['Error'])
cc[cc['Error']!=0].count().values[0]
```

```python
print('Total Error= {}%'.format(round((cc[cc['Error']!=0].count().values[0]/(len(df['RainTomorrow'])-1))*100,4)))
```

# Fill NaN number

```python
data_number['MinTemp']=data_number['MinTemp'].fillna(round(data_number['MinTemp'].mean(),1))
data_number['MaxTemp']=data_number['MaxTemp'].fillna(round(data_number['MaxTemp'].mean(),1))
data_number['WindGustSpeed']=data_number['WindGustSpeed'].fillna(data_number['WindGustSpeed'].median())
data_number['WindSpeed9am']=data_number['WindSpeed9am'].fillna(data_number['WindSpeed9am'].median())
data_number['WindSpeed3pm']=data_number['WindSpeed3pm'].fillna(data_number['WindSpeed3pm'].median())
data_number['Humidity9am']=data_number['Humidity9am'].fillna(round(data_number['Humidity9am'].median()))
data_number['Humidity3pm']=data_number['Humidity3pm'].fillna(round(data_number['Humidity3pm'].median()))
data_number['Pressure9am']=data_number['Pressure9am'].fillna(round(data_number['Pressure9am'].mean()))
data_number['Pressure3pm']=data_number['Pressure3pm'].fillna(round(data_number['Pressure3pm'].mean()))
data_number['Temp9am']=data_number['Temp9am'].fillna(round(data_number['Temp9am'].mean()))
data_number['Temp3pm']=data_number['Temp3pm'].fillna(round(data_number['Temp3pm'].mean()))
```

```python
data_number.info()
```

```python
for i in data_number.columns.drop(['RainTomorrow','Date']):
    plt.figure(figsize=(10,6))
    sns.boxplot(data = data_number,x='RainTomorrow',y=i)
    plt.show()
```
